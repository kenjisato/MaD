```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tsd = pd.read_stata("~/Data/10sd_jan15_2014/10SD_jan15.dta")
```

```python
tsd2 = (tsd[(tsd.Year > 1979) & (tsd.Year < 2013) & (tsd.Variable != "VA")]
        .drop(["Region"], axis=1)
        .set_index(["Country", "Regioncode", "Year"]))
tsd2
```

```python
def per_emp(x):
    numerator = x.loc[x["Variable"] != "EMP", "AGR":"SUM"]
    denominator = x.loc[x["Variable"] == "EMP", "AGR":"SUM"]
    return (numerator / denominator)
```

```python
tsd2.groupby(["Country", "Regioncode", "Year"]).get_group(('ARG', 'LAM', 2005))
```

```python
productivity = (pd.concat({ k : per_emp(v) for k, v in tsd2.groupby(["Country", "Regioncode", "Year"])})
                .reset_index()
                .drop(columns=["level_0", "level_1", "level_2"])
                .set_index(["Country", "Regioncode", "Year"]))
productivity
```
